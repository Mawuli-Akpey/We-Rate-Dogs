## 1.3 Method 2 (json file from Twitter API)

```python
# Query Twitter API for each tweet in the Twitter archive and save JSON in a text file
# These are hidden to comply with Twitter's API terms and conditions
consumer_key = 'HIDDEN'
consumer_secret = 'HIDDEN'
access_token = 'HIDDEN'
access_secret = 'HIDDEN'
```

```python
auth = OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)
```

```python
api = tweepy.API(auth, wait_on_rate_limit=True)
```

```python
# NOTE TO STUDENT WITH MOBILE VERIFICATION ISSUES:
# df_1 is a DataFrame with the twitter_archive_enhanced.csv file. You may have to
# change line 17 to match the name of your DataFrame with twitter_archive_enhanced.csv
# NOTE TO REVIEWER: this student had mobile verification issues so the following
# Twitter API code was sent to this student from a Udacity instructor
# Tweet IDs for which to gather additional data via Twitter's API
tweet_ids = twitter_archive.tweet_id.values
len(tweet_ids)
```

```python
# Query Twitter's API for JSON data for each tweet ID in the Twitter archive
count = 0
fails_dict = {}
start = timer()
# Save each tweet's returned JSON as a new line in a .txt file
with open('tweet-json.txt', 'w') as outfile:
    # This loop will likely take 20-30 minutes to run because of Twitter's rate limit
    for tweet_id in tweet_ids:
        count += 1
        print(str(count) + ": " + str(tweet_id))
        try:
            tweet = api.get_status(tweet_id, tweet_mode='extended')
            print("Success")
            json.dump(tweet._json, outfile)
            outfile.write('\n')
        except tweepy.TweepyException as e:
            print("Fail")
            fails_dict[tweet_id] = e
            pass
end = timer()
print(end - start)
print(fails_dict)
```

```python
tweets_list = []
with open('tweet-json.txt','r') as file:
    for line in file:
        data = json.loads(line)
        tweets_list.append(data)
```

## 6.2 Which algorithim has the best predicition rate?

```python
#We first calculate the mean for each algorithm and store it in a variable
algor_1_mean = twitter_clean.algor_1_conf.mean()
algor_2_mean = twitter_clean.algor_2_conf.mean()
algor_3_mean = twitter_clean.algor_3_conf.mean()

algor_1_mean, algor_2_mean, algor_3_mean
```

```python
#We use matplotlib here to generate a pie chart showing the percentage mean each algorithm comprises
plt.pie([algor_1_mean, algor_2_mean, algor_3_mean], labels=('Algorithm 1', 'Algorithm 2', 'Algorithm 3'),
        explode = [0.05,0,0.1], autopct='%1.1f%%')
plt.title('Percentage Algorithm Success')
plt.show()
```

From this pie chart it is evident that Algorithm 1 had the best prediction percentage by far

We already notice that the most popular stage is pupper and the least popular is floofer

## 6.4 What were the highest ratings? How were the different dog stages represented in each of these ratings?

First of all, we'll study the distribution of the rating_numerator column

```python
sns.displot(twitter_clean, x= 'rating_numerator', hue='stage_y')
plt.title('Distrubition of Ratings');
```

We can see that from a rating value of 21 upwards, we don't have significant representation for the other rating values

We'll confirm this with a count plot


```python
plt.figure(figsize=(13,6))
sns.countplot(x='rating_numerator', data=twitter_clean, color=base_color)
plt.title('Count of Ratings');
```

Confirmed!

Let's now focus on the top 10 rating values

```python
d = twitter_clean['rating_numerator'].value_counts()[:10]
plt.figure(figsize=(10,5))
bx = sns.countplot(x='rating_numerator', data=twitter_clean[twitter_clean['rating_numerator'].isin(d.index)])
for container in bx.containers:
    bx.bar_label(container)
plt.title('Top 10 Highest Ratings');
```

Our top 10 highest ratings ranged from between 11 and ending at 20 with most people giving a 12/10 rating

Now we determine how the different dog stages fared amongst these ratings

```python
#Over here, we simply repeat the code and add a hue for stages
d = twitter_clean['rating_numerator'].value_counts()[:10]
plt.figure(figsize=(15,8))
sns.countplot(x='rating_numerator', hue='stage_y', data=twitter_clean[twitter_clean['rating_numerator'].isin(d.index)])
plt.title('Top 10 Ratings by Dog Stage');
```

Pupper clearly had the most rating at each stage 8 times ouf 10.
